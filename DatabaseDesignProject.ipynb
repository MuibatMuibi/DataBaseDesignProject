# **Adventure Work Analysis**



# **Introduction**

The aim of this project is to show the steps to analyse the activities of the company between the year 2020-2022. This starts from the data definition language stage and Data manipulation stage in SSMS to reporting in SSRS. This analyses is focused on different categores discussed below to ensure increased revenue/profit generation, improved customers satisfaction and reduced return rate. The analysis has been done using the below parameters:

##

-  The Revenue/Profit generation by products category, territory and Year: To help understand where the business is doing well and where improvements are necessary. 
-  The Highest grossing customer: This would help in understanding our customers as it is only when we udnerstand them that we can provide them with unmatched customer satisfaction taht would eventually improve our bottomlines
-  The return rate: High return would mean defective products or lack of customers' fulfillment and eventual loss of revenue. This analysis will give an insight into the products with the highest retune, and this could help in identifying any undetected issue in the product production.

##  **The Dataset**

The dataset contains the sales and return tables between the year 2020-2022. Both tables contains the measurable information ablot the company such as sales, orderd quantity, and quantity returned. It also has the dimension tables calendar, customers, product, product categories, product subcategories and Territory.

## **Data Source**

The data [AdventureWorks](https://www.kaggle.com/datasets/algorismus/adventure-works-in-excel-tables) used for this project was sourced from kaggel.

##  Fact Table

Two fact tables named Sales and Return tables are used for this project.

-  **Sales Fact Table:** This contains the measurable and quantifiable data useful for analysis and reporting on the key perfoemance metrics of Adventure works such as Quantity sold, revenue and profit. This table establishes foreign key relationships to related dimension tables.

-  **Return Fact Table:** This table contains the quantity of items returned after sales and useful for analysis on product defects and quality control analysis. This table establishes foreign key relationships to related dimension tables.

-  The next step is to identify the columns to be included in the Territory table such as (Territorykey, Region, Country and Continent).This is important as it gives the system order to arrnage the column.
-  The SELECT statement gets the column bnames form the stging tabl using an alias t.
-  FROM indicates the source table which in this case is the stg.Territory Lookup.
- The selected columns must tally with the specified listing in the Insert to statement. And this ensures that only the necessary columns are retirved from the stg.Products Lookup table.
-  Data Validation uses the the WHERE clause filters the Territory, ensuring that only valid and unique products are inserted. And it contains 2 key conditions whicha re to filter out Non-Numeric products key if any exists and to avoid duplicates.
-  Go is the batch seperator

-  This query populates the f.Sales fact table from the staging table stg.[qSales Data].
-  The target table is the f.Sales and the columns are specified for insertion of values. Columns such as OrderDate, ordernNumber,ProductKey,CustomerKey,territoryKey,OrderQuantity, Retail Price and revenue.
-  These column names are selected using an alias s.
-  FROM statement indicates the source to come from the staging table.

    WHILE @Date <= @EndDate

            @Date,

        );
